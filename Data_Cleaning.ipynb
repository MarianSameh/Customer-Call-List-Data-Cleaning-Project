Load the Data

```python
import pandas as pd 

df = pd.read_excel(r"C:\Users\mariansameh\Downloads\Customer Call List.xlsx") 
```

```python
df.head()
```

Check data types

```python
df.dtypes
```

Remove unnecessary data 

```python
df=df.drop('Not_Useful_Column', axis=1)
```

```python
df=df.drop_duplicates()
```

Standardize Data

```python
df['Last_Name']=df['Last_Name'].str.strip('/')
df['Last_Name']=df['Last_Name'].str.strip('...')
df['Last_Name']=df['Last_Name'].str.strip('_')
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace(r'[-/|]', '', regex=True)
```

```python
df['Phone_Number'] = (
    df['Phone_Number'].str.slice(0, 3) + '-' +
    df['Phone_Number'].str.slice(3, 6) + '-' +
    df['Phone_Number'].str.slice(6, 10)
)
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
```

```python
Address Splitting
```

```python
df[['Street_Address', 'City', 'State_Zip']] = df['Address'].str.split(pat=',', n=2, expand=True)
df
```

drop rows with no phone number

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df
```

Reset index 

```python
df = df.reset_index(drop=True)
df
```

Handle null values 

```python
df=df.replace(['nan','N/a','Na--'],'')
df=df.fillna('')
```

Save the cleaned data into an excel file

```python
df.to_excel('C:\\Users\\mariansameh\\Downloads\\Customer Call List.xlsx', index=False, sheet_name='Cleaned Customer Call List')
```
